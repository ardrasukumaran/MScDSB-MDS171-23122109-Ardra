```python
import random

# Global list to store student data
students = []

# Function to calculate the grade based on the score
def calculate_grade(score):
    if score >100:
        return 'INVALID'
    elif score >= 90:
        return 'A+'
    elif score >= 80:
        return 'A'
    elif score >= 70:
        return 'B'
    elif score >= 60:
        return 'C'
    elif score >= 50:
        return 'D'
    else:
        return 'ERROR'

# Function to input student details and calculate grades
def student_grading_system():
    num_students = int(input("Enter the number of students:"))

    for i in range(num_students):
        student_name = input(f"Enter the name of student {i + 1}:")
        student_reg = input(f"Enter Student Register Number of {student_name}:")
        mark1 = random.randint(5, 25)
        mark2 = int(input("Enter Marks obtained out of 25 in subject 2:"))
        mark3 = int(input("Enter Marks obtained out of 25 in subject 3:"))
        mark4 = int(input("Enter Marks obtained out of 25 in subject 4:"))
        total = mark1 + mark2 + mark3 + mark4
        student_score = (total / 100) * 100

        grade = calculate_grade(student_score)

        student_data = {
            'Name': student_name,
            'Reg': student_reg,
            'mark1': mark1,
            'mark2': mark2,
            'mark3': mark3,
            'mark4': mark4,
            'Score': student_score,
            'Grade': grade
        }

        students.append(student_data)
        
     # Display the results
    print("\nStudent Grading Report:")
    print("--"*65)
    print("Name\t|\tReg No\t|\tRandom Mark 1\t|\tMark 2\t|\tMark 3\t|\tMark 4\t|\tScore\t|\tGrade\t|")
    print("--"*65)
    for student in students:
        print(f"{student['Name']}\t|\t{student['Reg']}\t|\t\t{student['mark1']}\t|\t{student['mark2']}\t|\t{student['mark3']}\t|\t{student['mark4']}\t|\t{student['Score']}\t|\t{student['Grade']}\t|\n")
    print("+"*130)


# Function to search for a student's grade by name
def search_student_grade(search_name):
    for student in students:
        if student['Name'] == search_name:
            return student['Grade']
    # Return None if the student is not found
    return None  

if __name__ == "__main__":
    # Function to input student details and calculate grades
    student_grading_system() 

    # Asking the user to enter a student name to search for
    search_name = input("\nEnter a student name to search for:")

    # Search for the student's grade by name
    grade = search_student_grade(search_name)

    # Print the result
    if grade is not None:
        print(f"The grade for {search_name} is {grade}")
    else:
        print(f"Student not found in the records.")

#To find the maximum Score
max_score = 0
for student in students:
    score = student['Score']
    if score > max_score:
        max_score = score
print("Maximum Score from the Class is:",max_score)

#To find the minimum score
min_score = 100
for student in students:
    score = student['Score']
    if score < max_score:
        min_score = score
print("Minimum Score from the Class is:",min_score)
```
