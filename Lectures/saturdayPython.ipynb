```python
#count the number odd and even numbers in the list and print the odd numbers count and even numbers count 
#along with the numbers

#You will have a function that will take a list of numbers, that returns the count of odd numbers and even numbers
#and the new list of odd numbers and even numbers.
def numbers(NumLst):
    Even = []
    odd = []
    for item in NumLst:
        if item %2 ==0:
            Even.append(item)
        else:
            odd.append(item)
    print("Even list is:",Even)
    print("Odd list is",odd)
    print("Count of odd is:",len(odd))
    print("count of Even is:",len(Even))

NumLst = []
n = int(input("Enter the number of numbers:"))
for i in range(n):
    inp = int(input("Enter numbers"))
    NumLst.append(inp)
print(NumLst)

numbers(NumLst)


```

```python
import random
import csv
# #class is the keyword. __init__ is a member function to initialise the data entries in the class.
# #self.name and self.students are data members/properties or attributes of the class

class MSCDSB:

    def __init__(self):
        self.name = "MSc DS:B"
        self.students = ["A","B","C"]
    
    def printStudents(): #Another member function
        for students in self.students:
            print(students)

    

# ##to have mltiple copies of the class we need to call the class
# #to access the object inside the class

obj = MSCDSB()
print(obj.name)
print(obj.students)
obj.printStudents()
```

```python
class grocery:

    def __init__(self, item_name, quantity):
        self.item = item_name
        self.qnty = quantity
        self.menu = {
            "potato" : 20,
            "apple" : 40,
            "bread" : 30

        }
    def printPrice(self): 
        print("Price of Grocery is:")
        print("-------------------------")
        print("Item Name:\t" , self.item)
        print("Quantity:\t", self.qnty)
        total = self.qnty*self.menu['apple']
        print("Total Price = \t", total)

order = grocery('apple',6)
order.printPrice()
    
    
```
