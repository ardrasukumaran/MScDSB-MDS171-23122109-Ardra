Write a python program with the following specifications:
There will be function named "create dataset" which will take a dictionary, filename, number of rows as arguments. 
The dictionary is having the following structure:
{
    "column name":[domain of values]
}
your function will select random values from each of the domain and create a row for you. the number of rows that has to be created is a parameter that has been taken. 
Save the file with the with the requitred number of rows
load this dataset using pandas and display basic details

```python
import random
def createDataset(dic1,fileName,rowNumber ): 
    file = open(fileName,"w+")
    file.write("category\t"+"name\t"+"age\t"+"gender\t"+"place\t"+"type\n")
    for i in range(0,rowNumber):
        category = random.choice(dic1["Category"])
        name = random.choice(dic1["Name"])
        age = random.choice(dic1["Age"])
        gender = random.choice(dic1["Gender"])
        place = random.choice(dic1["Place"])
        type = random.choice(dic1["Employment Type"])
        file.write(category+"\t"+ name+"\t"+age+"\t"+gender+"\t"+place+"\t"+type+"\n")


dic1 = {
        #"columnName":["Category","Name","Age","Gender","Place","Employment Type"],
        "Category":["1","2","3","4","5"],
        "Name": ["Ardra","Gokul","Riya PC","Devika","Aadith","Jofin","Nandhana"],
        "Age":["20","21","22"],
        "Gender":["Female","Male"],
        "Place":["Kannur","Palakkad","Lavasa","Pune"],
        "Employment Type": ["student","self employed","unemployed","employee"]
    }

rowNumber = int(input("Enter the number of rows: "))
fileName = input("Enter the name of the file: ")
createDataset(dic1,fileName,rowNumber)
```
