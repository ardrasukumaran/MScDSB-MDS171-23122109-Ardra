```python
List1 = [761, 123, 585, 275, 194, 840, 999, 622, 763, 40, 825, 687, 402, 338, 611, 987, 739, 95, 819, 568, 320, 126, 688, 948, 514, 223, 751, 379, 857, 767, 492, 162, 437, 934, 430, 343, 725, 511, 580, 678, 708, 823, 346, 179, 786, 663, 195, 890, 267, 502, 214, 300, 786, 961, 235, 559, 992, 919, 948, 318, 57, 787, 653, 317, 305, 172, 966, 182, 970, 17, 89, 624, 629, 595, 485, 752, 984, 978, 460, 124, 321, 256, 614, 852, 811, 561, 127, 207, 406, 615, 888, 391, 691, 470, 911, 401, 137, 600, 478, 682]

#Mean

def mean():
    Sum = 0
    Number = 0
    for i in List1:
        Sum = Sum + i
        Number = Number +1
    mean = Sum/Number
    return mean

meanVal = mean()
print("Mean: ",meanVal)


#Minimum
def minVal(List1):
    minVal = List1[1]
    for i in List1:
        if minVal<=i:
            minVal = minVal
        else:
            minVal = i
    return minVal

a=minVal(List1)
print("Minimum: ",a)


#Maximum

def maxVal(List1):
    maxVal = List1[1]
    for i in List1:
        if maxVal>=i:
            maxVal = maxVal
        else:
            maxVal = i
    return maxVal

b=maxVal(List1)
print("Maximum: ",b)

#Range
Range = b-a
print("Range: ",Range)

```
